```python
import requests

url = "https://api-core.caviarnine.com/v1.0/shapeliquidity/get_components"
response = requests.get(url)
components_data = response.json()
```

```python
import pandas as pd

# Extract the shapeliquidity list from components_data
shapeliquidity_list = components_data['shapeliquidity']

# Convert to DataFrame
df = pd.DataFrame(shapeliquidity_list)
# ensure total_amount_x and total_amount_y are floats
df['total_amount_x'] = df['total_amount_x'].astype(float)
df['total_amount_y'] = df['total_amount_y'].astype(float)

# Display the DataFrame
df
```

```python
df.columns
```

```python
# Get unique values in the 'status' column
print("\nUnique status values:")
print(df['status'].unique())

# Get value counts to see the distribution
print("\nStatus value counts:")
print(df['status'].value_counts())
```

```python
# Filter for specific tokens and select columns
filtered_tokens = ['xwBTC', 'LSULP', 'XRD', 'xETH', 'xUSDT','FLOOP','SRG']
# filtered_tokens = ['xwBTC']
columns_to_show = ['component_address', 'total_amount_x', 'total_amount_y', 
                   'token_x_name', 'token_x_symbol', 'token_y_name', 'token_y_symbol','status','is_active']

resource_address = 'resource_rdx1tknxxxxxxxxxradxrdxxxxxxxxx009923554798xxxxxxxxxradxrd'
# filtered_df = filtered_df[(df['token_x_address'] == resource_address) | (df['token_y_address'] == resource_address)]

filtered_df = df[
    ((df['token_x_address'] == resource_address) | (df['token_y_address'] == resource_address)) &
    (df['token_x_address'].notna() | df['token_y_address'].notna()) 
    # & (df['token_x_symbol'].isin(filtered_tokens)) 
    # & (df['token_y_symbol'].isin(filtered_tokens)) 
    & (df['is_active'] )
][columns_to_show]

filtered_df
```

```python
# Pools where token_y_name is 'Radix'
df_radix_y = filtered_df[filtered_df['token_y_name'] == 'Radix']
grouped_by_x = (
    df_radix_y
    .groupby('token_x_symbol')['total_amount_y']
    .sum()
    .reset_index()
    .rename(columns={'token_x_symbol': 'pooled_token', 'total_amount_y': 'xrd amount'})
        .sort_values('xrd amount', ascending=False)

    # .assign(**{'xrd amount': lambda df: df['xrd amount'].apply(lambda v: f"{v:.6f}")})
    .assign(dapp='caviarnine')

)
print("Sum of total_amount_x by token_x_name (where token_y_name is 'Radix'):")
# print(grouped_by_x.head(10))

# Pools where token_x_name is 'Radix'
df_radix_x = filtered_df[filtered_df['token_x_name'] == 'Radix']
grouped_by_y = (
    df_radix_x
    .groupby('token_y_symbol')['total_amount_x']
    .sum()
    .reset_index()
    .rename(columns={'token_y_symbol': 'pooled_token', 'total_amount_x': 'xrd amount'})
    .sort_values('xrd amount', ascending=False)
    .assign(dapp='caviarnine')

)
print("\nSum of total_amount_x by token_y_name (where token_x_name is 'Radix'):")
# print(grouped_by_y.head(10))
# Concatenate the two summary DataFrames
combined_df = pd.concat([grouped_by_x, grouped_by_y], ignore_index=True).sort_values('xrd amount', ascending=False).assign(**{'xrd amount': lambda df: df['xrd amount'].apply(lambda v: f"{v:.6f}")})


# Display the concatenated result
print("\nConcatenated grouped_by_x and grouped_by_y:")
print(combined_df.head(25))

```

```python
import requests

url = "https://api.ociswap.com/pools?cursor=0&limit=25"
response = requests.get(url)
response.raise_for_status()
ociswap_pool_raw_data = response.json()
ociswap_pool_data = pd.DataFrame(ociswap_pool_raw_data['data'])
ociswap_pool_data
```

```python
ociswap_pool_data[['token_x_name', 'token_y_name']] = ociswap_pool_data['name'].str.split('/', expand=True)
ociswap_pool_data
```

```python
# Extract nested fields from the x and y JSON columns
ociswap_pool_data['x_token_symbol'] = ociswap_pool_data['x'].apply(
    lambda j: j.get('token', {}).get('symbol') if isinstance(j, dict) else None
)
ociswap_pool_data['x_total_value_locked_now'] = (
    ociswap_pool_data['x']
    .apply(lambda j: j.get('total_value_locked', {}).get('token', {}).get('now') 
           if isinstance(j, dict) else None)
    .astype(float)
)
ociswap_pool_data['y_token_symbol'] = ociswap_pool_data['y'].apply(
    lambda j: j.get('token', {}).get('symbol') if isinstance(j, dict) else None
)

ociswap_pool_data['y_total_value_locked_now'] = (
    ociswap_pool_data['y']
    .apply(lambda j: j.get('total_value_locked', {}).get('token', {}).get('now') 
           if isinstance(j, dict) else None)
    .astype(float)
)


# Optional: view the augmented DataFrame
ociswap_pool_data.head()
```

```python
# collect unique symbols from both columns and exclude 'XRD'
symbols = pd.unique(ociswap_pool_data[['x_token_symbol', 'y_token_symbol']].values.ravel())
# non_xrd_symbols = [s for s in symbols if s != 'XRD']
# non_xrd_symbols
```

```python

filtered_tokens = ['XRD','xwBTC','xETH','xUSDT','FLOOP','SRG','xUSDC']
filtered_ociswap_pool_data = ociswap_pool_data[
    ((ociswap_pool_data['x_token_symbol'] == 'XRD') | (ociswap_pool_data['y_token_symbol'] == 'XRD')) 
    & (ociswap_pool_data['x_token_symbol'].isin(symbols)) 
    & (ociswap_pool_data['y_token_symbol'].isin(symbols)) 
]

filtered_ociswap_pool_data.head()
filtered_ociswap_pool_data[filtered_ociswap_pool_data['x_token_symbol'] == 'OCI']
```

```python
# Filter out rows where x_token_symbol is 'XRD'
non_xrd_pools = filtered_ociswap_pool_data[filtered_ociswap_pool_data['x_token_symbol'] != 'XRD']

# Group by the non‐XRD x_token_symbol and sum up y_total_value_locked_now
x_token_summary = (
    non_xrd_pools
    # .groupby('x_token_symbol', as_index=False)['y_total_value_locked_now']
    .groupby('x_token_symbol')['y_total_value_locked_now']
    .sum()
    .reset_index()

    .rename(columns={'x_token_symbol': 'pooled_token', 'y_total_value_locked_now': 'xrd amount'})
    .sort_values('xrd amount', ascending=False)

    # .drop(columns=['index'])
    .assign(**{'xrd amount': lambda df: df['xrd amount'].apply(lambda v: f"{v:.6f}")})
    .assign(dapp='ociswap')
)


# View the result
x_token_summary
```
